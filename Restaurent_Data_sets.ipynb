```python
a='Amount After Tax'
b=Data.groupby("Client Name").sum().sort_values('Amount After Tax',ascending=False)[['Amount After Tax']][0:50]
sns.barplot(x=a,y=b.index,data=b)
```

```python
c='Amount After Tax'
d=Data.groupby("Product Name").sum().sort_values('Amount After Tax',ascending=True)[['Amount After Tax']][0:25]
sns.barplot(x=c,y=d.index,data=d)
```

```python
e='Quantity'
f=Data.groupby('Product Name').sum().sort_values('Quantity',ascending=False)[['Quantity']][0:25]
sns.barplot(x=e,y=f.index,data=f)
```

```python
y=Data.groupby('Client Name').sum().sort_values('Product Name',ascending=False)[['Product Name']][0:25]
sns.barplot(x='Product Name',y=y.index,data=y)
```

```python
y=Data.groupby('Date').sum().sort_values('',ascending=False)[[]][0:25]
sns.barplot(x=,y=y.index,data=y)
```

```python
from wordcloud import WordCloud
```

```python
Text=' '.join(Data['Product Name'].dropna().astype(str).values)
```

```python
wordCloud=WordCloud(background_color='White',width=700,height=400).generate(Text)
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.imshow(wordCloud,interpolation='bilinear')
ax.set_title('Word Cloud of Product Name Column')
ax.set_xticks([])
ax.set_yticks([])
```
